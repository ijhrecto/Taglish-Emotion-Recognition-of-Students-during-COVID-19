```python
import io
import pandas as pd
df2 = pd.read_csv(io.BytesIO(uploaded['cleaned_data_for_translation (1).csv']))
# Dataset is now stored in a Pandas Dataframe

df2.count
df2
```

```python
#join emoji and text
import numpy as np
twts = []

for i in df2.index:
  if pd.isnull(df2.loc[i, "emoticons"]): 
      twts.append(df2.loc[i, "text"]) 

  else:
    temp = str(df2.loc[i, "text"]).split()
    temp_emoj = str(df2.loc[i, "emoticons"]).split(',')
    temp_ind = [int(k) for k in (str(df2.loc[i,"emoticons_index"]).split(','))]

    for j, k in zip(temp_ind, temp_emoj):
      temp.insert(j,k)
    
    word = " ".join(temp)
    twts.append(word)
```

```python
twt_ls = []
temp = []

for i in df3["text"]:
  temp.append(str(i))
  if not (4000 > len('+'.join(temp))):
    twt_ls.append(temp)
    temp = []

df4 = pd.DataFrame(data = {'translated_text':['+'.join(i) for i in twt_ls]})
```

```python
from google_trans_new import google_translator  
translator = google_translator() 

for ind in df4.index:
  translate_text = translator.translate(df4.loc[ind, "translated_text"],lang_src='tl', lang_tgt='en')
  df4.loc[ind, "translated_text"] = translate_text
```

```python
translated_df = pd.DataFrame(columns =['twts'])
translated_df
```

```python
#separate tweets into rows
import re
#delimiters = "++", "+ +"
#regexPattern = '|'.join(map(re.escape, delimiters))

for row in df4['translated_text']:
  strlist = row.split('+')
  for s in strlist:
    new_row = {'twts' : s}
    translated_df = translated_df.append(new_row, ignore_index = True)
```
